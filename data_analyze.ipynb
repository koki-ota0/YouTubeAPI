```python
import os
import json
import random
import requests
import numpy as np
import pandas as pd
from googleapiclient.discovery import build
from googleapiclient.errors import HttpError
import networkx as nx
import matplotlib.pyplot as plt
```

```python
def read_json(file):
    # JSONファイルを読み込む
    with open(file, 'r') as f:
        data = json.load(f)

    # DataFrameに変換する
    df = pd.DataFrame(data)
    return df

file_path = "data/comments/@tsensei_comments.json"
comments = read_json(file_path)
```

```python
random_sample = comments #.sample(n=1000, random_state=42)

video_ids = set()
commenter_ids = set()
edges = []

for index, row in random_sample.iterrows():
    v = row['snippet']['videoId']
    c = row['snippet']['topLevelComment']['snippet']['authorDisplayName']
    if v and c:
        video_ids.add(v)
        commenter_ids.add(c)
        edges.append((v, c))
        
print(f"ビデオの数：{len(video_ids)}")
print(f"コメント者の数：{len(commenter_ids)}")

# 2部グラフを作成
B = nx.Graph()
B.add_nodes_from(video_ids, bipartite=0)  # 動画のIDを1つの集合に割り当てる
B.add_nodes_from(commenter_ids, bipartite=1)  # コメント投稿者のIDをもう1つの集合に割り当てる

# エッジを追加
B.add_edges_from(edges)

# ノードの色を設定
node_colors = ['skyblue' if node in video_ids else 'lightpink' for node in B.nodes]

# フラクタルレイアウトでノードを配置
pos = nx.fruchterman_reingold_layout(B)

# グラフを可視化
plt.figure(figsize=(10, 6))
nx.draw(B, pos, with_labels=False, node_color=node_colors, node_size=500)
plt.title("Bipartite Graph (Fruchterman-Reingold Layout)")
plt.show()
```
